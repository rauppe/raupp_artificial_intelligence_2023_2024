<h1> Racism proof </h1>

```python
#another solution
race_white = (df['Suspect-ethnicity'] == 'White - English/Welsh/Scottish/Northern Irish/British' )| \
            (df['Suspect-ethnicity'] == 'White - Any other White background') | (df['Suspect-ethnicity'] == 'White - Irish')
```

```python
race_white.sum()
```

```python
race_known = (df['Suspect-ethnicity'] != 'Other ethnic group - Not stated') & (~ df['Suspect-ethnicity'].isnull())
```

```python
race_known.sum()
```

```python
race_white.sum() * 100 / race_known.sum()
```

```python
unnec = df[df['Outcome'] == 'A no further action disposal']
```

```python
race_known_unnec = (unnec['Suspect-ethnicity'] != 'Other ethnic group - Not stated') & (~ unnec['Suspect-ethnicity'].isnull())
```

```python
race_white.sum() * 100 / race_known_unnec.sum()
```
